```python
headerName = ['ID', 'Range', 'Value', 'Machine Type', 'Zip Code']
myDF = pd.read_csv('/content/test.csv', header=None, names=headerName)
print("The first way to select Value column: ")
print(myDF['Value'].head())
print()
print("The second way to select Value column: ")
print(myDF.Value.head())
```

### Answer: 
  Use **Bracket Notation** is more OOP because: 
    (1) Bracket can always works when column name includes a space
    (2) Column name matches a DataFrame method
    (3) Column name matches a Python keyword
    (4) Column name is stored in a variable
    (5) Column name is an integer
